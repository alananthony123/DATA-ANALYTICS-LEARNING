```python
def corr_coeff(df3, x ,y):
    x_bar = df3[x].mean()
    y_bar = df3[y].mean()
    x_sd = df3[x].std()
    y_sd = df3[y].std()
    total_prod = (((df3[x]-x_bar)/x_sd)*((df3[y]-y_bar)/y_sd)).sum()
    corr = total_prod/(df3.shape[0]-1)


    return corr
```
